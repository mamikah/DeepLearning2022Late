# matplotlibグラフ表示

## matplotlibの読み込み

```python
# 読込命令
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## データの準備と描画

```python
t = np.arange(0.0,2.0,0.01)
# データの加工
s = 1 + np.sin(2 * np.pi * t)

# グラフの描画
fig,ax = plt.subplots()
ax.plot(t,s)
# グラフにラベルを設定する
ax.set(xlabel = 'time(s)',ylabel = 'voltage(mv)',title = 'About as simple as it gets,folks')
# グラフにグリッドを付ける
ax.grid()
ax.plot(t,s)
plt.show(ax)
```

## ラベルとグリッド
